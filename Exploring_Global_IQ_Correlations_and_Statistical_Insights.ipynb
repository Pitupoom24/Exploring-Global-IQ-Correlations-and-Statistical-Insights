**Choose and explain at least one question you have about your dataset that you proposed in Section 4. Clearly state the questions you have chosen and how you plan on using Python to answer them. I choose this question: what education system/learning strategy is favored most for high average IQ. I assume that the mean years of schooling is based on the education system in a given country because most countries have a mandatory minimum requirement for basic education. I will use the one-tailed two-sample hypothesis test to determine whether there is a difference between the average IQ of countries whose schooling category is between 10 and 15 years and the average IQ of countries whose schooling category is less than or equal to 10 years.

```python
df["Mean years of schooling - 2021"].max().item()
```

```python
mean_0_to_5 = df[df["Mean years of schooling - 2021"] < 5]["Average IQ"].mean()
mean_5_to_10 = df[( df["Mean years of schooling - 2021"] >= 5 ) & ( df["Mean years of schooling - 2021"] < 10 )]["Average IQ"].mean()
mean_10_to_15 = df[( df["Mean years of schooling - 2021"] >= 10 ) & ( df["Mean years of schooling - 2021"] < 15 )]["Average IQ"].mean()

print(f"The mean between 0 and 5 is {mean_0_to_5:.2f}")
print(f"The mean between 5 and 10 is {mean_5_to_10:.2f}")
print(f"The mean between 10 and 15 is {mean_10_to_15:.2f}")
```

**State the Hypotheses**
- `Null Hypothesis`: The average IQ of the countries whose schooling category is between 10 and 15 years = The average IQ of the countries whose schooling category is less than 10 years.
- `Alternative Hypothesis`: The average IQ of the countries whose schooling category is between 10 and 15 years > The average IQ of the countries whose schooling category is less than 10 years.

```python
df_10_to_15 = df[( df["Mean years of schooling - 2021"] >= 10 ) & ( df["Mean years of schooling - 2021"] < 15 )]
df_less_than_10 = df[df["Mean years of schooling - 2021"] < 10]

from statsmodels.stats.weightstats import ztest
result = ztest(x1=df_10_to_15["Average IQ"].dropna(), x2=df_less_than_10["Average IQ"].dropna(), value=0, alternative='larger')
result
```

Since p-value = 1.3715e-24 < 0.05, we **reject** the claim saying that the average IQ of the countries whose schooling category is between 10 and 15 years = The average IQ of the countries whose schooling category is less than 10 years since the observed difference between two groups is too big to be due to chance.
